# **data preprocessing/data cleaning & Feature Engineering**

As by the above insights we get to know 77% of the Cabin Column Is represented by Null values[Cloumn will be dropped]
Null values in Age can be replaced by mean/average 
Nulls in Embarked will be dropped 
also 

```python
titanic.drop(['PassengerId','Cabin' ,'Ticket','Name'],axis = 1 , inplace = True)
```

```python
titanic
```

```python
titanic.Age.fillna(titanic.Age.mean(),inplace = True)
```

```python
titanic['Age'].isnull().sum()
```

```python
titanic.dropna(inplace = True)
titanic.shape #2-rows-dropped-bcoz-of-2nulls-in-embarked_column
```

```python
titanic['Sex'] = titanic['Sex'].map({'male':0,'female':1})
titanic['Sex']
```

```python
#converting_Categorical_variable_into-Numerical_values_____-one-hot-encoding/get_dummies
ports = pd.get_dummies(titanic.Embarked, prefix = 'Embarked')
ports
```

```python
titanic.drop(['Embarked'],axis = 1, inplace = True)
titanic = titanic.join(ports)
```

```python
titanic #now-entire-data-isin-numerical_form
```

# **splitting-training&model-fitting**

```python
X = titanic.drop(['Survived'], axis = 1)
y = titanic.Survived.copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test ,y_train , y_test = train_test_split(X,y , test_size = 0.2 , random_state = 7)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train , y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
